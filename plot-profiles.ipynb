```python
mode = "cpu"

branch1 = "master"
time_master_cpu = np.loadtxt(f"Profiling/{branch1}_time_cpu.txt")
data_master_cpu = np.loadtxt(f"Profiling/{branch1}_memory_cpu.txt")
if mode == "gpu":
    time_master_gpu = np.loadtxt(f"Profiling/{branch1}_time_gpu.txt")
    data_master_gpu = np.loadtxt(f"Profiling/{branch1}_memory_gpu.txt")

branch2 = "pr"
time_pr_cpu = np.loadtxt(f"Profiling/{branch2}_time_cpu.txt")
data_pr_cpu = np.loadtxt(f"Profiling/{branch2}_memory_cpu.txt")
if mode == "gpu":
    time_pr_gpu = np.loadtxt(f"Profiling/{branch2}_time_gpu.txt")
    data_pr_gpu = np.loadtxt(f"Profiling/{branch2}_memory_gpu.txt")
```

```python
plt.plot(time_master_cpu, data_master_cpu-min(data_master_cpu), "b--", label="master CPU", linewidth=2)
plt.plot(
    time_pr_cpu, data_pr_cpu - min(data_pr_cpu), "r--", label="pr CPU", linewidth=1
)
if mode == "gpu":
    plt.plot(
        time_master_gpu, data_master_gpu, "b", label="Before", linewidth=2
    )
    plt.plot(time_pr_gpu, data_pr_gpu, "orange", label="After", linewidth=1)
plt.xlabel("Time (s)", fontsize=15)
plt.ylabel("Memory Usage (MB)", fontsize=15)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.savefig(f"memory-reduction{mode}.png", dpi=500)
plt.show()
```

```python
branch3 = "pr"
time_pr0_cpu = np.loadtxt(f"Profiling/{branch3}-before-gc_time_cpu.txt")
data_pr0_cpu = np.loadtxt(f"Profiling/{branch3}-before-gc_memory_cpu.txt")
time_pr0_gpu = np.loadtxt(f"Profiling/{branch3}-before-gc_time_gpu.txt")
data_pr0_gpu = np.loadtxt(f"Profiling/{branch3}-before-gc_memory_gpu.txt")
```
