```python
import os
import pickle

import matplotlib.pyplot as plt
```

```python
cases = [
    "jax060",
    "jax053",
    "jax052",
    "jax051",
    "jax050",
    "jax0438",
    "jax0434",
    "jax0424",
]
data = []
for case in cases:
    with open(f"{case}.pickle", "rb") as f:
        data.append(pickle.load(f))

name = "test_proximal_freeb_jac_batched"
for i in range(len(data)):
    plt.plot(data[i][name]["t"], data[i][name]["mem"], label=cases[i])
plt.title(name)
plt.ylabel("Memory Usage [MB]")
times = [data[i][name]["t"][-1] for i in range(len(data))]
max_time = max(times) + 0.5
plt.xlabel("Time [s]")
plt.xlim([0, max_time])
plt.grid(True)
plt.legend()
plt.tight_layout()
png = "compare.png"
plt.savefig(png, dpi=150)
```
